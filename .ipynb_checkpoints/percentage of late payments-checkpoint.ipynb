```python
#standard importing stuff
import pandas as pd
import numpy as np
import datetime
```

```python
#read dataset
datalog=pd.read_csv('BPI Challenge 2018.csv', sep=';')
```

```python
#explore
print(datalog.head())
print(datalog.info())
```

```python
#print first event
datalog[datalog['Case ID']==datalog['Case ID'][0]]

```

```python
#only keep the relevant attributes
datalog=datalog[['Case ID', 'Activity', 'Complete Timestamp']]
```

```python
#convert timestamp to datetime objects (ignore error)
datalog['Complete Timestamp']=pd.to_datetime(datalog['Complete Timestamp'], format='%Y/%m/%d')
```

```python
datalog['Payment In Time']=False
datalog['BeginExists'] = False 
datalog['CaseStarted'] = datetime.datetime(2000, 1, 1)
datalog['LastBeginPayment'] = datetime.datetime(2000, 1, 1)
datalog['LastAbortPayment'] = datetime.datetime(2000, 1, 1)
```

```python
caseidlist=datalog['Case ID'][datalog['Activity']=='Payment application-Application-begin payment']
```

```python
datalog2 = datalog.groupby('Case ID')
```

```python
datalog2[datalog2['BeginExists'] == True][datalog2['Activity']=='Payment application-Application-begin payment'].sort_values().tail(n=1)
```

```python
caseidlist.drop_duplicates()

datalog["BeginExists"][tuple(datalog['Case ID'].values) in caseidlist]=True
datalog['LastBeginPayment'][datalog['BeginExists']==True]=(datalog["Complete Timestamp"][datalog['Activity']=='Payment application-Application-begin payment'].sort_values()).tail(n=1)

```

```python
for caseid in datalog["Case ID"]:

    datalog[datalog["Case ID"] == caseid]['LastBeginPayment'][datalog['BeginExists']==True]=(datalog[datalog["Case ID"] == caseid]["Complete Timestamp"][datalog['Activity']=='Payment application-Application-begin payment'].sort_values()).tail(n=1)
```
