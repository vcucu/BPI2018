```python
#only keep the relevant attributes
datalog=datalog[['Case ID', 'Activity', 'Complete Timestamp']]
```

```python
#convert timestamp to datetime objects (ignore error)
datalog['Complete Timestamp']=pd.to_datetime(datalog['Complete Timestamp'], format='%Y/%m/%d')
```

```python
datalog['Payment In Time']=range(len(datalog))

for caseid in datalog["Case ID"]:
    caseidlist=[]
    if caseid not in caseidlist:
        startyear=0
        endyear=1
        if len([datalog['Complete Timestamp'][(datalog['Case ID']==caseid) & (datalog['Activity']== 'Payment application-Application-mail income')][0]])>0:
            startyear=datalog['Complete Timestamp'][(datalog['Case ID']==caseid) & (datalog['Activity']== 'Payment application-Application-mail income')][0].year
        if len([datalog['Complete Timestamp'][(datalog['Case ID']==caseid) & (datalog['Activity']== 'Payment application-Application-begin payment')][0]])>0:
            beginpaymentyear=datalog['Complete Timestamp'][(datalog['Case ID']==caseid) & (datalog['Activity']== 'Payment application-Application-begin payment')][0].year
        datalog['Payment In Time'][datalog['Case ID']==caseid] = startyear==beginpaymentyear
        caseidlist.append(caseid)
```
